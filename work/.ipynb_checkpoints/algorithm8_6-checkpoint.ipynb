```julia
function runEMpath(u0,T,Nref,d,m,fhandle,ghandle,kappa,M)
    s = 0.0
    Mstep = 1000
    m0 = 1

    for mm in 1:Mstep:M

        MM = min(Mstep, M - mm + 1)
        u00 = u0[:, mm:m0+MM-1]

        #save rng state
        rng = Random.default_rng()
        savedState = copy(rng)

        #reference solution
        tref, uref = EMpath(u00, T, Nref, d, m,
                            fhandle, ghandle, 1, MM)

        #reset
        Random.seed!(rng, savedState.seed)

        #approximated solution
        t, u = EMpath(u00, T, Nref, d, m,
                      fhandle, ghandle, kappa, MM)

        #error at the last point
        err = u[:, :, end] .- uref[:, :, end]

        S += sum(err .* err)
        m0 += MM
    end

    rmsErr = sqrt(S / M)

    return rmsErr, t, u, tref, uref
end
        
    
```
